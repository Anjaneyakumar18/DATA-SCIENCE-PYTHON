```python
# Logistic Regression is a supervised learning algorithm used for binary classification,
# predicting probabilities based on input features using the sigmoid function.
# It estimates the relationship between independent variables and a categorical outcome
```

```python
# Importing libraries and modules 
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, confusion_matrix, precision_score, recall_score, roc_auc_score
import matplotlib.pyplot as plt
```

```python
# Data defination
X=np.array([3,7,6,5,2,1,7,8,9,4,3,11,12,5]).reshape(-1,1)
Y=np.array([0,1,1,1,0,0,1,1,1,0,0,1,1,1])
```

```python
X
```

```python
Y
```

```python
# Train test Split
x_train,x_test,y_train,y_test=train_test_split(X,Y)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predicted_y=model.predict(x_test)
```

```python
predicted_y
```

```python
# Precision: Measures the proportion of correctly predicted positive cases out of all predicted positives.
# Recall: Measures the proportion of correctly predicted positive cases out of all actual positives.
# F1 Score: Harmonic mean of precision and recall, balancing both metrics.
# ROC AUC: Evaluates the model's ability to distinguish between classes by measuring the area under the ROC curve.
# Confusion Matrix: A table showing the counts of true positives, true negatives, false positives, and false negatives.
```

```python
confussion_mat=confusion_matrix(y_test,predicted_y)
```

```python
sns.heatmap(confussion_mat)
plt.title("Confusion Matrix")
plt.xlabel("Prediceted")
plt.ylabel("True")
plt.show()
```

```python
precision=precision_score(y_test,predicted_y)
```

```python
recall=recall_score(y_test,predicted_y)
```

```python
f1=f1_score(y_test,predicted_y)
```

```python
roc_auc=roc_auc_score(y_test,predicted_y)
```

```python
print(f'Precision : {precision}')
print(f'Recall : {recall}')
print(f'F1_score : {f1}')
print(f"ROC_AUC : {roc_auc}")
```
