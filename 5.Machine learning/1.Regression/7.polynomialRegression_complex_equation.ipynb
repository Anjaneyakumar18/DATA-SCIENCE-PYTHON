```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.linear_model as lm
import sklearn.preprocessing as sp
```

```python
# working with more complex polynomial equation
# (3x**3)+(x**2)+(7x)+2
```

```python
data={
    'x':[i for i in range(20)],
    'y':[(3*(i**3))+(i**2)+(7*i)+2 for i in range(20)]
}
```

```python
df=pd.DataFrame(data)
```

```python
#Relationship between X and Y
```

```python
plt.plot(data['x'],df['y'],label="Relationship between X and Y Training Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
```

```python
#using Linear Regression alone
```

```python
model=lm.LinearRegression()
```

```python
model.fit(pd.DataFrame({'x':df['x']}),df.y)
```

```python
pre=[]
for i in data['x']:
    val=model.predict(pd.DataFrame({'x':[i]}))
    pre.append(val[0])
```

```python
# ploting Actual vs Predicted
```

```python
plt.plot(df.x,df.y,label="X vs Y Actual from Training Data",color="Green")
plt.plot(df.x,pre,label="X vs Y Predicted from Training Data",color="Red")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Actual Y vs predicted Y with Linear regression")
plt.grid(True)
```

```python
plt.plot(data['x']+[40],data['y']+[194022],label="X vs Y Actual One test data point included",color="Green")
plt.plot(data['x']+[40],pre+[model.predict([[40]])[0]],label="X vs Y Predicted test data predicted",color="Red")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Actual Y vs predicted Y with Linear regression (with 1 test data far to actual points)")
plt.grid(True)
```

```python
X=df[['x']]
Y=df.y
```

```python
poly=sp.PolynomialFeatures(degree=3)
```

```python
poly_x=poly.fit_transform(X)
```

```python
model=lm.LinearRegression()
```

```python
model.fit(poly_x,df.y)
```

```python
pre=model.predict(poly_x)
```

```python
# ploting after Polynomial featuring
```

```python
plt.scatter(df.x,df.y,color="Red",label="Actual X - Y relationship")
plt.plot(df.x,pre,color="Green",label="Predicted X - Y relationship")
plt.legend()
plt.xlabel('X')
plt.ylabel('y')
plt.grid(True)
```
