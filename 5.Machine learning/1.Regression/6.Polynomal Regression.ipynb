```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sna
from pandas import DataFrame
```

```python
data={
    'x':[i for i in range(20)],
    'y':[i**2 for i in range(20)]
}
```

```python
df=pd.DataFrame(data)
```

```python
plt.plot(df.x,df.y,label="Actual")
plt.xlabel("x",fontsize="16")
plt.ylabel('y',fontsize="16")
plt.title("X and Y relationship in Train data")
plt.legend()
```

```python
model=LinearRegression()
```

```python
model.fit(df[['x']],df['y'])
```

```python
pre=[]
for i in range(20):
    val=model.predict(pd.DataFrame({'x':[i]}))
    pre.append(val[0])
```

```python
plt.plot(df.x,df.y,label="X vs Y Actual",color="Green")
plt.plot(df.x,pre,label="X vs Y Predicted",color="Red")
plt.legend()
plt.xlabel("X",fontsize="16")
plt.ylabel("Y",fontsize="16")
```

```python
plt.plot(data['x']+[40],data['y']+[1600],label="X vs Y Actaul")
plt.plot(data['x']+[40],np.append(pre,model.predict(pd.DataFrame({'x':[40]}))[0]),label="X vs Y Predicted in Test Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
```

```python
data = {
    'x': [i for i in range(20)],
    'y': [i**2 for i in range(20)]
}
```

```python
df = pd.DataFrame(data)
```

```python
X = df[['x']]  
y = df['y']   
```

```python
degree = 2
poly = PolynomialFeatures(degree=degree, include_bias=False)
X_poly = poly.fit_transform(X) 
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
y_pred = model.predict(X_poly)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual values', s=50)
plt.plot(X, y_pred, color='red', label='Polynomial Fit (degree={})'.format(degree))
plt.title("Prediction vs Actual Values")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
```
