```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data={
    "Age":[x for x in range(22,45)],
    "Salary":[34,38,36,42,46,44,54,58,56,61,67,63,70,75,72,79,83,81,87,82,84,91,95]
}
```

```python
dataset=pd.DataFrame(data)
```

```python
model=LinearRegression()
```

```python
model.fit(dataset[['Age']],dataset['Salary'])
```

```python
predicted=model.predict(dataset[['Age']])
```

```python
plt.scatter(dataset['Age'],dataset['Salary'],label="Actual Values")
plt.plot(dataset['Age'],predicted,color="orange",label="Predicted values")
plt.legend()
plt.xlabel("Age (years)")
plt.ylabel("Salary (₹ in Thousands/Month)")
plt.title("Simple Linear Regression: Age vs Salary")
```

```python
unknown_data=pd.DataFrame({
    "Age":[x for x in range(45,55)]
})
```

```python
unknown_data_predictions=model.predict(unknown_data)
```

```python
plt.scatter(unknown_data,unknown_data_predictions,label="Predictions")
plt.plot(unknown_data,unknown_data_predictions,color="red",label="Regression line")
plt.legend()
plt.xlabel("Age in years",fontsize="16")
plt.ylabel("Salary in Thousands/month",fontsize="16")
```
