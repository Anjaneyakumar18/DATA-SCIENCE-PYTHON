```python
dataset=pd.DataFrame(data)
```

```python
dataset.replace({"Domain":{
    "AI":1,
    "ML":2,
    "Data Science":3,
    "WebDev":4,
    "MobileDev":5
}},inplace=True)
```

```python
domain_wise={}
for i in range(115):
    if data['Domain'][i] in domain_wise:
        domain_wise[data['Domain'][i]].append(data['Salary'][i])
    else:
        domain_wise[data['Domain'][i]]=[]
```

```python
for key in domain_wise.keys():
    domain_wise[key]=sum(domain_wise[key])/len(domain_wise[key])
```

```python
plt.bar(domain_wise.keys(),domain_wise.values(),color=['red','orange','blue','green','yellow'])
plt.title("Domain vs Salary")
```

```python
age_wise={}
for i in range(115):
    if data['Age'][i] in age_wise:
        age_wise[data['Age'][i]].append(data['Salary'][i])
    else:
        age_wise[data['Age'][i]]=[]
```

```python
plt.plot(age_wise.keys(),age_wise.values())
plt.xlabel('Age')
plt.ylabel("Salary")
plt.title("Age vs Salary")
```
