```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating three DataFrames with different relationships
df = pd.DataFrame({
    "one": [x for x in range(100)],          # Linear relationship: y = x
    "two": [y * y for y in range(100)]       # Polynomial relationship: y = x^2
})

df2 = pd.DataFrame({
    "x": [i for i in range(100)],            # Linear relationship: y = x + 10
    "y": [j + 10 for j in range(100)]
})

df3 = pd.DataFrame({
    "x": [i for i in range(100)],            # Linear relationship: y = 2x
    "y": [j * 2 for j in range(100)]
})

# Creating subplots (2x2 grid)
fig, ax = plt.subplots(2, 2, figsize=(10, 10))  # Set figure size for better visibility

# Top left: df["one"] vs. df["two"]
ax[0][0].plot(df["one"], df["two"], label="y=x^2")
ax[0][0].set_title("Polynomial Relationship (y = x^2)")
ax[0][0].legend()

# Top right: df["two"] vs. df["one"]
ax[0][1].plot(df["two"], df["one"], label="Inverse Polynomial")
ax[0][1].set_title("Inverse Polynomial")
ax[0][1].legend()

# Bottom left: df2['x'] vs. df2['y']
ax[1][0].plot(df2['x'], df2['y'], label="y=x+10", color="green")
ax[1][0].set_title("Linear Relationship (y = x + 10)")
ax[1][0].legend()

# Bottom right: df3['x'] vs. df3['y']
ax[1][1].plot(df3['x'], df3['y'], label="y=2x", color="orange")
ax[1][1].set_title("Linear Relationship (y = 2x)")
ax[1][1].legend()

# Show the plots
plt.tight_layout()
plt.show()

# Comments on Relationships:
# - Linear relationship: Straight-line graphs (y = x, y = x + 10, y = 2x) as seen in df2 and df3.
#   These graphs show a constant rate of change, which means the ratio between y and x is fixed.
#
# - Polynomial relationship: Curved graph (y = x^2 in df) exhibits a quadratic increase.
#   In this case, as x grows, y grows at an accelerating rate (non-linear growth).
```
